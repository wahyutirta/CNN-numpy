```python
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
def load_train_details(**params):
    method = params.get("method", "adam")
    epochs = params.get("epochs", 500)
    batch = params.get("learningRate", 32)
    learningRate = params.get("batch", 0.01)
    mainPath = params.get("mainPath", "")
    stringName = str("-" + "jepun" + str(batch) + str(learningRate) + str(method) + str(epochs))

        
    fname = "valid_history" + str(stringName) + ".npz"
    path = os.path.join(mainPath,str(method),fname)
    arr_files = np.load(path)
    valid_loss_history, valid_acc_history, valid_ep = arr_files['arr_0'], arr_files['arr_1'], arr_files['arr_2']
    #print("validation====")
    #print(valid_loss_history.shape)
    #print(valid_acc_history.shape)
    #print(valid_steps.shape)
        
    fname = "step_history" + str(stringName) + ".npz"
    path = os.path.join(mainPath,str(method),fname)
    arr_files = np.load(path)
    loss_history, acc_history, weights_history, n_steps = arr_files['arr_0'], arr_files['arr_1'], arr_files['arr_2'], arr_files['arr_3']
    
    #print("step====")
    #print(loss_history.shape)
    #print(acc_history.shape)
    #print(weights_history.shape)
    #print(n_steps.shape)    
    
    fname = "epoch_history" + str(stringName) + ".npz"
    path = os.path.join(mainPath,str(method),fname)
    arr_files = np.load(path)
    epoch_loss, epoch_acc, epoch_weight, epochs = arr_files['arr_0'], arr_files['arr_1'], arr_files['arr_2'], arr_files['arr_3']
    #print("epoch====")
    #print(epoch_loss.shape)
    #print(epoch_acc.shape)
    #print(epoch_weight.shape)
    #print(epochs.shape)
    
    return valid_loss_history, valid_acc_history, valid_ep, loss_history, acc_history, weights_history, n_steps, epoch_loss, epoch_acc, epoch_weight
```

```python
#mainPath = os.path.dirname(os.path.abspath(__file__)) #file path main.py
mainPath = globals()['_dh'][0]
workPath = os.path.split(mainPath) #path working folder (whole file project)
imagePath = "data_jepun"

method = "adam"
epochs = 101
batch = 32
learningRate = 0.001

adam_valid_loss_history, adam_valid_acc_history, adam_valid_ep, adam_loss_history, adam_acc_history, adam_weights_history, adam_n_steps, adam_epoch_loss, adam_epoch_acc, adam_epoch_weight = load_train_details(mainPath=mainPath,epochs=epochs,method=method, batch=batch, learningRate=learningRate )
epochs_array = np.array(list(range(epochs)))

method = "rmsprop"
epochs = 101
batch = 32
learningRate = 0.001

rms_valid_loss_history, rms_valid_acc_history, rms_valid_ep, rms_loss_history, rms_acc_history, rms_weights_history, rms_n_steps, rms_epoch_loss, rms_epoch_acc, rms_epoch_weight = load_train_details(mainPath=mainPath,epochs=epochs,method=method, batch=batch, learningRate=learningRate )
epochs_array = np.array(list(range(epochs)))
```

```python
def plotting(epochs_array, array1, array2, tittle, label1, label2, y_label):
    plt.plot(epochs_array, array1, label = label1)

 
    plt.plot(epochs_array, array2, label = label2)
  
    # naming the x axis
    plt.xlabel('epochs')
    # naming the y axis
    plt.ylabel(y_label)
  
    # giving a title to my graph
    plt.title(tittle)

    # show a legend on the plot
    plt.legend()
  
    # function to show the plot
    plt.show()
```
